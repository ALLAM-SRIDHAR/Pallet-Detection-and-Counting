```python
#defining the model
from ultralytics import YOLO
model1=YOLO("yolov8n.pt")
#running the model
results1=model1.train(data='/content/-Pallets-Detection-and-Counting-1/data.yaml',epochs = 50,imgsz=640)
model1.val()
```

```python
# Load the best trained model
model1 = YOLO("/content/runs/detect/train/weights/best.pt")  # Adjust path if needed
```

```python
results1 = model1.predict(
    source="/content/-Pallets-Detection-and-Counting-1/test/images/10_jpg.rf.896da99ba02074b4b23198e0181c7cdc.jpg",
    conf=0.25,
    save=False
)

# Extract and count class IDs
boxes = results1[0].boxes
class_ids = boxes.cls.cpu().numpy()

# Get class names
names = model1.names

# Count detections per class
import numpy as np
unique, counts = np.unique(class_ids, return_counts=True)
detection_counts = {
    names[int(cls_id)]: int(count)  # Convert np.int64 to regular int
    for cls_id, count in zip(unique, counts)
}

print(detection_counts)
```

```python
# -------------------------------
# Display the image with bounding boxes
# -------------------------------
import cv2
import matplotlib.pyplot as plt

# Get the rendered image (with bounding boxes)
rendered_img = results1[0].plot()  # returns a BGR NumPy array

# Convert BGR to RGB
rgb_img = cv2.cvtColor(rendered_img, cv2.COLOR_BGR2RGB)

# Show image using matplotlib
plt.figure(figsize=(10, 8))
plt.imshow(rgb_img)
plt.axis("off")
plt.title("Detected Objects with Bounding Boxes")
plt.show()
```

```python
# Run validation and get the metrics object
metrics = model1.val()  # Run validation

# Extract key metrics (precision, recall, mAP)
precision = metrics.box.p  # Precision for each class
recall = metrics.box.r     # Recall for each class
f1_score = metrics.box.f1  # F1 score for each class
map50 = metrics.box.map50  # mAP at IoU=0.50
map50_95 = metrics.box.map  # mAP at IoU=0.5:0.95

# Print the values
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1_score}")
print(f"mAP@0.5: {map50}")
print(f"mAP@0.5:0.95: {map50_95}")
```
